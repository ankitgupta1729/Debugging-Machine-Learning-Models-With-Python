## Section: Debugging in software development
Here you can see different version of a fucntion called `multiply()` with different issues that could result in eitehr error,when runnign the code, or wrong (unexpected) output.

The following code is the correct one that returns multiplication of two numbers.

```python
# Correct code
def multiply(x, y):
  z = x * y
  return z

print(multiply(2, 3))
```

The following code will return an error as the indentation of the line `z = x * y` in the function is not correct.

```python
# Code with indentation issue
def multiply(x, y):
z = x * y
return z

print(multiply(2, 3))
```

The following code will return an error as `retunr` is not an identifiable syntax by Python.

```python
# Code with typo
def multiply(x, y):
  z = x * y
  retunr z

print(multiply(2, 3))
```

The following code will return an incorrect (unexpected) answer of 8 instead of 6 as `**` is used instead of `*` for multiplying the input variables of `multiply()` functions.

```python
# Incorrect use of ** for multiplication
def multiply(x, y):
  z = x ** y
  return z

print(multiply(2, 3))
```
