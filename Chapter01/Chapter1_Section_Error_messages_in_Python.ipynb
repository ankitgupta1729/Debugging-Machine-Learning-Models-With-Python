We first want to understand how wrong code indentation could result in incorrect answer while the code would run without returning an error. The following code shows `odd_counter()` that calculates number of odd numbers in an input list of numbers, which is three considering the `num_list`.

```python
#Intended code
def odd_counter(num_list: list):
  """
  :param num_list: list of integers to be checked for identifying odd numbers

  :return: return an integer as the number of odd numbers in the input list
  """
  odd_count = 0
  for num in num_list:
    if (num % 2) == 0:
      print("{} is even".format(num))
    else:
      print("{} is even".format(num))
      odd_count += 1

  return odd_count


num_list = [1, 2, 5, 8, 9]
print(f'Total number of odd numbers in the list: {odd_counter(num_list)}')
```

Now we change indentation of `odd_count += 1` which results in no error but returns total number of numbers in the input `num_list` instead of only odd numbers.

```python
#Unintended code
def odd_counter(num_list: list):
  """
  :param num_list: list of integers to be checked for identifying odd numbers

  :return: return an integer as the number of odd numbers in the input list
  """
  odd_count = 0
  for num in num_list:
    if (num % 2) == 0:
      print("{} is even".format(num))
    else:
      print("{} is even".format(num))
    odd_count += 1

  return odd_count

num_list = [1, 2, 5, 8, 9]
print(f'Total number of odd numbers in the list: {odd_counter(num_list)}')
```

Next, we see how `a.isnull()` and `a.isnan()` returns **AttributeError**. \\
First we can see that `a.isnull()` returns **AttributeError** as 'list' object has no attribute 'isnull'.

```python
a = [1, 2]
a.isnull()
```

Also `a.isnan()` returns **AttributeError** in the following code as 'float' object has no attribute 'isnan'.

```python
import numpy as np
a = np.nan
a.isnan()
```
