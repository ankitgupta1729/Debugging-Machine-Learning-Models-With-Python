## Section: Visualizations could be misleading

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
time_range = np.arange(0, 200)
series1 = [iter+100*(np.random.random()-0.5) for iter in time_range]
series2 = [2]+series1[0:-1]

plt.plot(time_range, series1)
plt.plot(time_range, series2)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
plt.xlabel('time', fontsize = 11)
plt.ylabel('target variable', fontsize = 11)

plt.show()
```
