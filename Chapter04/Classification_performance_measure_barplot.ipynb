## Section: Label-based performance metrics

```python
import matplotlib.pyplot as plt
import pandas as pd

# create data
df = pd.DataFrame([['0.2', 0.8, 0.8, 0.02, 1, 0.89, 0.5, 0.02],
                   ['0.4', 0.6, 0.6, 0.02, 1, 0.75, 0.5, 0.02],
                   ['0.6', 0.4, 0.4, 0.02, 1, 0.57, 0.5, 0.02],
                   ['0.8', 0.2, 0.2, 0.02, 1, 0.33, 0.5, 0.02]],
                  columns=['Real positive fraction', 'acc', 'precision',
                           'recall', 'specificity', 'F1',
                           'balanced acc', 'mcc'])
# view data
print(df)

# plot grouped bar chart
df.plot(x='Real positive fraction',
        ylabel = 'performance',
        fontsize = 11,
        kind='bar',
        stacked=False,
        title='')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.show()
```
