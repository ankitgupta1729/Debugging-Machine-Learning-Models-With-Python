## Section: Outlier removal
Visualizing parts of the diabetes dataset from scikit-learn showcasing outliers in the data.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

diabetes_frame = load_diabetes(as_frame=True)['data']

bmi = diabetes_frame.loc[:,'bmi']
plt.hist(bmi, bins=15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('Body Mass Index (BMI)', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.show()


plt.boxplot(bmi)
plt.xticks([1], [''])
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylabel('Body Mass Index (BMI)', fontsize = 14)
plt.show()
```
