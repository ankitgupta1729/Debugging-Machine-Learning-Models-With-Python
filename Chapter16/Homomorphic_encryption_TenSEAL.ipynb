```python
import tenseal as ts

context = ts.context(ts.SCHEME_TYPE.BFV, poly_modulus_degree=4096, plain_modulus=1032193)
context
```

```python
plain_list = [50, 60, 70]
# encrypting the vector of integers into a BFVVector, a vector type that uses the BFV scheme
encrypted_list = ts.bfv_vector(context, plain_list)
print("Encrypted vector of size:", encrypted_list.size())
encrypted_list
```

```python
add_result = encrypted_list + [1, 2, 3]
print(add_result.decrypt())
```

```python
sub_result = encrypted_list - [1, 2, 3]
print(sub_result.decrypt())
```
